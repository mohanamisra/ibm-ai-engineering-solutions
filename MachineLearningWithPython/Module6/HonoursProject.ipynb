```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./Weather_Data.csv')
df.head()
```

```python
df.replace(['No', 'Yes'], [0, 1], inplace=True)
df.head()
```

```python
print(df.dtypes)
```

```python
# CLEARLY, WindDir9am, WindDir3pm and WindGustDir are 3 categorical values that need to be converted by one hot encoding
df = pd.get_dummies(df, columns=['WindGustDir', 'WindDir9am', 'WindDir3pm'])
df.head()
```

```python
df.drop('Date', axis=1, inplace=True)
```

```python
df = df.astype('float64')
df.head()
```

```python
columns = df.columns
len(columns)
```

```python
x = df[[column for column in columns if column != 'RainTomorrow']]
x.head()
```

```python
y = df['RainTomorrow']
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
predictions = []
```

```python
# LINEAR REGRESSION CODE

from sklearn.linear_model import LinearRegression

lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train, y_train)
lin_reg_pred = lin_reg_model.predict(x_test)
predictions.append(lin_reg_pred)

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, lin_reg_pred)
mae = mean_absolute_error(y_test, lin_reg_pred)
r2 = r2_score(y_test, lin_reg_pred)

results = pd.DataFrame({
    'Metrics': ['Mean Squared Error', 'Mean Absolute Error', 'R2 Score'],
    'Values': [mse, mae, r2],
})
print(results)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=4)
knn_model.fit(x_train, y_train)
knn_pred = knn_model.predict(x_test)
predictions.append(knn_pred)

from sklearn.metrics import accuracy_score, jaccard_score, f1_score
acc = accuracy_score(y_test, knn_pred)
jacc = jaccard_score(y_test, knn_pred)
f1 = f1_score(y_test, knn_pred)
knn_results = pd.DataFrame({
    'Metrics': ['Accuracy', 'Jaccard Score', 'F1 Score'],
    'Values': [acc, jacc, f1],
})
print(knn_results)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc_model = DecisionTreeClassifier(max_depth=4)
dtc_model.fit(x_train, y_train)
dtc_pred = dtc_model.predict(x_test)
predictions.append(dtc_pred)

tree_acc = accuracy_score(y_test, dtc_pred)
tree_jacc = jaccard_score(y_test, dtc_pred)
tree_f1 = f1_score(y_test, dtc_pred)
dtc_results = pd.DataFrame({
    'Metrics': ['Accuracy', 'Jaccard Score', 'F1 Score'],
    'Values': [tree_acc, tree_jacc, tree_f1],
})
print(dtc_results)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

logreg_model = LogisticRegression(solver='liblinear')
logreg_model.fit(x_train, y_train)
logreg_pred = logreg_model.predict(x_test)
predictions.append(logreg_pred)
logreg_predict_proba = logreg_model.predict_proba(x_test)
predictions_proba = []
predictions_proba.append(logreg_predict_proba)

accuracy = accuracy_score(y_test, logreg_pred)
jacc = jaccard_score(y_test, logreg_pred)
f1 = f1_score(y_test, logreg_pred)
logreg_log_loss = log_loss(y_test, logreg_predict_proba)

log_reg_results = pd.DataFrame({
    'Metrics': ['Accuracy', 'Jaccard Score', 'F1 Score', 'Log Loss'],
    'Values': [accuracy, jacc, f1, logreg_log_loss]
})
print(log_reg_results)
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='poly', gamma='auto')
svm_model.fit(x_train, y_train)
svm_pred = svm_model.predict(x_test)
predictions.append(svm_pred)
svm_acc = accuracy_score(y_test, svm_pred)
svm_jacc = jaccard_score(y_test, svm_pred)
svm_f1 = f1_score(y_test, svm_pred)
svm_results = pd.DataFrame({
    'Metrics': ['Accuracy', 'Jaccard Score', 'F1 Score'],
    'Values': [svm_acc, svm_jacc, svm_f1],
})
print(svm_results)
```

```python
results['Model'] = 'Linear Regression'
knn_results['Model'] = 'KNN'
dtc_results['Model'] = 'Decision Tree'
log_reg_results['Model'] = 'Logistic Regression'
svm_results['Model'] = 'SVM'

results = results[['Model', 'Metrics', 'Values']]
knn_results = knn_results[['Model', 'Metrics', 'Values']]
dtc_results = dtc_results[['Model', 'Metrics', 'Values']]
log_reg_results = log_reg_results[['Model', 'Metrics', 'Values']]
svm_results = svm_results[['Model', 'Metrics', 'Values']]

report = pd.concat([results, knn_results, dtc_results, log_reg_results, svm_results], ignore_index=True)

print(report)
```
